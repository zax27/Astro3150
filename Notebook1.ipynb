```python
hol=np.loadtxt('holland_temperature.dat')
newy=np.loadtxt('newyork_temperature.dat')
beij=np.loadtxt('beijing_temperature.dat')
plt.plot(np.linspace(1, 12, 12),hol)
plt.plot(np.linspace(1, 12, 12),newy)
plt.plot(np.linspace(1, 12, 12),beij)
plt.xlabel('Number of the month')
plt.ylabel('Mean monthly temperature(C)')
plt.xticks(np.linspace(1, 12, 12))
plt.legend(['Holland','New York','Beijing'], loc='best');
```

```python
a=np.loadtxt('holland_temperature.dat') 
s=np.loadtxt('holland_seawater.dat')
plt.subplot(211)
plt.plot(a,'b',label='Air temp.')
plt.plot(s,'r',label='Sea temp.')
plt.legend(loc='best')
plt.ylabel('Temp. (C)')
plt.xlim(0, 11)
plt.xticks([])
plt.subplot(212)
plt.plot(a-s, 'ko')
plt.xticks(np.linspace(0, 11, 12),
           ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
plt.xlim(0, 11)
plt.ylabel('Air - Sea temp (C)');
```

```python
a=np.loadtxt('holland_temperature.dat') 
s=np.loadtxt('holland_seawater.dat')

plt.fill_between(range(1,13),a,color='g',alpha=0.3)
plt.fill_between(range(1,13),s,color='b',alpha=0.3)

plt.xticks(np.linspace(0,11,12), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug', 'Sep',' Oct','Nov','Dec'])
plt.xlim(1, 12)
plt.ylim(0, 20)
plt.xlabel('Month')
plt.ylabel('Temp.(C)');
```
